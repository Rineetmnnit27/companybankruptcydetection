```python


import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
from imblearn.over_sampling import SMOTE
```

```python
# ==========================================================
# Features and Target
# ==========================================================

X = df.drop("Bankrupt?", axis=1)

y = df["Bankrupt?"]

print(X.shape)
print(y.shape)
```

```python
constant_features = [
    col
    for col in X.columns
    if X[col].nunique()==1
]

print("Constant Features")

print(constant_features)
```

```python
X = X.drop(
    columns=constant_features
)
```

```python
selector = VarianceThreshold(
    threshold=0.0001
)

selector.fit(X)

selected_columns = X.columns[
    selector.get_support()
]

removed_columns = list(
    set(X.columns) -
    set(selected_columns)
)

print("Removed Features")

print(removed_columns)
```

```python
outlier_summary = []

for col in X.columns:

    q1 = X[col].quantile(0.25)

    q3 = X[col].quantile(0.75)

    iqr = q3-q1

    lower = q1 - 1.5*iqr

    upper = q3 + 1.5*iqr

    count = (
        (X[col]<lower) |
        (X[col]>upper)
    ).sum()

    outlier_summary.append(
        {
            "Feature":col,
            "Outliers":count,
            "Percentage":
            round(
                count/len(X)*100,
                2
            )
        }
    )

outlier_df = pd.DataFrame(
    outlier_summary
)

outlier_df.sort_values(
    "Percentage",
    ascending=False
).head(20)
```

```python
X_cap = X.copy()

for col in X.columns:

    q1 = X[col].quantile(0.25)

    q3 = X[col].quantile(0.75)

    iqr = q3-q1

    lower = q1-1.5*iqr

    upper = q3+1.5*iqr

    X_cap[col] = X[col].clip(
        lower,
        upper
    )

X = X_cap.copy()

print("Outlier Capping Completed")
```

```python


X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)

print(X_train.shape)

print(X_test.shape)
```

```python
print(y_train.value_counts())

print()

print(y_test.value_counts())
```

```python
plt.figure(figsize=(5,4))

sns.countplot(x=y_train)

plt.title("Training Class Distribution")

plt.show()
```

```python
smote = SMOTE(
    random_state=42
)

X_train_smote, y_train_smote = smote.fit_resample(

    X_train,

    y_train

)

print("Before SMOTE")

print(y_train.value_counts())

print()

print("After SMOTE")

print(y_train_smote.value_counts())
```

```python
train_df = X_train_smote.copy()

train_df["Bankrupt?"] = y_train_smote

train_df.to_csv(
    "train_processed.csv",
    index=False
)

test_df = X_test.copy()

test_df["Bankrupt?"] = y_test.values

test_df.to_csv(
    "test_processed.csv",
    index=False
)

print("Processed datasets saved successfully.")
```

```python


import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import RobustScaler

from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
# ==========================================================
# Load Processed Data
# ==========================================================

train = pd.read_csv("train_processed.csv")

test = pd.read_csv("test_processed.csv")

print(train.shape)

print(test.shape)
```

```python
X_train = train.drop("Bankrupt?", axis=1)

y_train = train["Bankrupt?"]

X_test = test.drop("Bankrupt?", axis=1)

y_test = test["Bankrupt?"]
```

```python
skewness = X_train.skew()

skewness.sort_values(ascending=False).head(20)
```

```python
plt.figure(figsize=(10,8))

skewness.sort_values().tail(20).plot(
    kind="barh"
)

plt.title("Most Skewed Features")

plt.show()
```

```python
high_skew = skewness[
    skewness.abs()>1
].index

print("Highly Skewed Features")

print(len(high_skew))
```

```python
pt = PowerTransformer(
    method="yeo-johnson"
)

X_train[high_skew] = pt.fit_transform(
    X_train[high_skew]
)

X_test[high_skew] = pt.transform(
    X_test[high_skew]
)
```

```python
for col in X_train.columns:
    print(col)
```

```python
liquidity = [
    " Current Ratio",
    " Quick Ratio",
    " Cash Ratio"
]

available = [c for c in liquidity if c in X_train.columns]

if available:

    X_train["Liquidity_Score"] = X_train[
        available
    ].mean(axis=1)

    X_test["Liquidity_Score"] = X_test[
        available
    ].mean(axis=1)
```

```python
profit = [

" ROA(C) before interest and depreciation before interest",

" Operating Gross Margin",

" Net profit before tax"

]

available = [c for c in profit if c in X_train.columns]

if available:

    X_train["Profitability_Score"] = X_train[
        available
    ].mean(axis=1)

    X_test["Profitability_Score"] = X_test[
        available
    ].mean(axis=1)
```

```python
leverage = [

" Debt ratio %",

" Liability to Equity"

]

available = [c for c in leverage if c in X_train.columns]

if available:

    X_train["Leverage_Score"] = X_train[
        available
    ].mean(axis=1)

    X_test["Leverage_Score"] = X_test[
        available
    ].mean(axis=1)
```

```python
efficiency = [

" Total Asset Turnover",

" Inventory Turnover Rate (times)",

" Receivable Turnover"

]

available = [c for c in efficiency if c in X_train.columns]

if available:

    X_train["Efficiency_Score"] = X_train[
        available
    ].mean(axis=1)

    X_test["Efficiency_Score"] = X_test[
        available
    ].mean(axis=1)
```

```python
if (
    " Debt ratio %" in X_train.columns
    and
    " Current Ratio" in X_train.columns
):

    X_train["Debt_Current"] = (

        X_train[" Debt ratio %"]

        /

        (X_train[" Current Ratio"]+1e-6)

    )

    X_test["Debt_Current"] = (

        X_test[" Debt ratio %"]

        /

        (X_test[" Current Ratio"]+1e-6)

    )
```

```python

if (
    " Net Value Per Share (B)" in X_train.columns
    and
    " Revenue Per Share (Yuan ¥)" in X_train.columns
):

    X_train["Revenue_to_NetWorth"] = (

        X_train[" Revenue Per Share (Yuan ¥)"]

        /

        (X_train[" Net Value Per Share (B)"]+1e-6)

    )

    X_test["Revenue_to_NetWorth"] = (

        X_test[" Revenue Per Share (Yuan ¥)"]

        /

        (X_test[" Net Value Per Share (B)"]+1e-6)

    )
```

```python
scaler = RobustScaler()

X_train = pd.DataFrame(

    scaler.fit_transform(X_train),

    columns=X_train.columns

)

X_test = pd.DataFrame(

    scaler.transform(X_test),

    columns=X_test.columns

)
```

```python
rf = RandomForestClassifier(

    n_estimators=300,

    random_state=42,

    n_jobs=-1

)

rf.fit(X_train,y_train)
```

```python
importance = pd.DataFrame({

"Feature":X_train.columns,

"Importance":rf.feature_importances_

})

importance = importance.sort_values(

"Importance",

ascending=False

)

importance.head(20)
```

```python
plt.figure(figsize=(10,8))

sns.barplot(

data=importance.head(20),

x="Importance",

y="Feature"

)

plt.title("Top 20 Important Features")

plt.show()
```

```python
selector = SelectFromModel(

rf,

threshold="median",

prefit=True

)

X_train_selected = selector.transform(

X_train

)

X_test_selected = selector.transform(

X_test
)

selected_columns = X_train.columns[
selector.get_support()
]

X_train = pd.DataFrame(

X_train_selected,

columns=selected_columns

)

X_test = pd.DataFrame(

X_test_selected,

columns=selected_columns
)

print(X_train.shape)

print(X_test.shape)
```

```python
train_final = X_train.copy()

train_final["Bankrupt?"] = y_train.values

train_final.to_csv(

"train_engineered.csv",

index=False

)

test_final = X_test.copy()

test_final["Bankrupt?"] = y_test.values

test_final.to_csv(

"test_engineered.csv",

index=False

)

print("Feature Engineering Completed Successfully")
```

```python


import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error
)

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
#this is asynthetic dataset generated for forecasting

np.random.seed(42)

dates = pd.date_range(
    start="2019-01-01",
    periods=72,
    freq="MS"
)

trend = np.linspace(50000,120000,72)

seasonality = 8000*np.sin(np.arange(72)*2*np.pi/12)

noise = np.random.normal(0,3000,72)

revenue = trend + seasonality + noise

revenue_df = pd.DataFrame({

    "Date":dates,

    "Revenue":revenue

})

revenue_df.head()
```

```python
plt.figure(figsize=(15,5))

plt.plot(

    revenue_df["Date"],

    revenue_df["Revenue"],

    linewidth=3

)

plt.title("Monthly Revenue")

plt.xlabel("Date")

plt.ylabel("Revenue")

plt.show()
```

```python
revenue_df["Month"] = revenue_df["Date"].dt.month

revenue_df["Quarter"] = revenue_df["Date"].dt.quarter

revenue_df["Year"] = revenue_df["Date"].dt.year
```

```python
revenue_df["Lag_1"] = revenue_df["Revenue"].shift(1)

revenue_df["Lag_3"] = revenue_df["Revenue"].shift(3)

revenue_df["Lag_6"] = revenue_df["Revenue"].shift(6)
```

```python
revenue_df["Rolling_Mean_3"] = (

    revenue_df["Revenue"]

    .rolling(3)

    .mean()

)

revenue_df["Rolling_STD_3"] = (

    revenue_df["Revenue"]

    .rolling(3)

    .std()

)
```

```python
revenue_df = revenue_df.dropna()

revenue_df.head()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(revenue_df["Revenue"])

print("ADF Statistic :", result[0])
print("p-value :", result[1])

if result[1] < 0.05:
    print("The series is stationary.")
else:
    print("The series is NOT stationary.")
```

```python
revenue_df["Revenue_diff"] = revenue_df["Revenue"].diff()

revenue_df = revenue_df.dropna()
```

```python
result = adfuller(revenue_df["Revenue_diff"])

print("ADF Statistic :", result[0])
print("p-value :", result[1])

if result[1] < 0.05:
    print("The differenced series is stationary.")
else:
    print("The differenced series is still NOT stationary.")
```

```python
plt.figure(figsize=(14,5))

plt.plot(

    revenue_df["Date"],

    revenue_df["Revenue_Diff"]

)

plt.title("Differenced Revenue")

plt.show()
```

```python
train_size = int(

    len(revenue_df)*0.8

)

train = revenue_df.iloc[:train_size]

test = revenue_df.iloc[train_size:]

print(train.shape)

print(test.shape)
```

```python
model = ARIMA(

    train["Revenue"],

    order=(2,1,2)

)

model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast = model_fit.forecast(

    steps=len(test)

)
```

```python
mae = mean_absolute_error(

    test["Revenue"],

    forecast

)

rmse = np.sqrt(

    mean_squared_error(

        test["Revenue"],

        forecast

    )

)

mape = np.mean(

    np.abs(

        (

            test["Revenue"]-

            forecast

        )

        /

        test["Revenue"]

    )

)*100

print("MAE :",mae)

print("RMSE :",rmse)

print("MAPE :",mape)
```

```python
plt.figure(figsize=(15,6))

plt.plot(

    train["Date"],

    train["Revenue"],

    label="Training"

)

plt.plot(

    test["Date"],

    test["Revenue"],

    label="Actual"

)

plt.plot(

    test["Date"],

    forecast,

    label="Forecast"

)

plt.legend()

plt.title("Revenue Forecast")

plt.show()
```

```python
future = model_fit.forecast(

    steps=12

)

future_dates = pd.date_range(

    revenue_df["Date"].max()+pd.offsets.MonthBegin(),

    periods=12,

    freq="MS"

)

future_df = pd.DataFrame({

    "Date":future_dates,

    "Forecasted_Revenue":future

})

future_df
```

```python
plt.figure(figsize=(15,6))

plt.plot(

    revenue_df["Date"],

    revenue_df["Revenue"],

    label="Historical Revenue"

)

plt.plot(

    future_df["Date"],

    future_df["Forecasted_Revenue"],

    color="red",

    linewidth=3,

    label="Forecast"

)

plt.legend()

plt.title("Future Revenue Forecast")

plt.show()
```

```python
future_df.to_csv(

    "forecasted_revenue.csv",

    index=False

)

print("Forecast saved successfully.")
```

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier

from sklearn.model_selection import (
    cross_val_score,
    StratifiedKFold
)

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    PrecisionRecallDisplay,
    classification_report
)

import joblib
```

```python
train = pd.read_csv("train_engineered.csv")

test = pd.read_csv("test_engineered.csv")
```

```python
X_train = train.drop("Bankrupt?", axis=1)

y_train = train["Bankrupt?"]

X_test = test.drop("Bankrupt?", axis=1)

y_test = test["Bankrupt?"]

print(X_train.shape)

print(X_test.shape)
```

```python
models = {

    "Logistic Regression":
        LogisticRegression(max_iter=500),

    "Decision Tree":
        DecisionTreeClassifier(
            random_state=42
        ),

    "Random Forest":
        RandomForestClassifier(
            n_estimators=300,
            random_state=42
        ),

    "XGBoost":
        XGBClassifier(
            random_state=42,
            eval_metric="logloss"
        )

}
```

```python
cv = StratifiedKFold(

    n_splits=5,

    shuffle=True,

    random_state=42

)

for name, model in models.items():

    scores = cross_val_score(

        model,

        X_train,

        y_train,

        scoring="roc_auc",

        cv=cv,

        n_jobs=-1

    )

    print()

    print(name)

    print("-"*40)

    print("Mean ROC-AUC:", scores.mean())

    print("Std ROC-AUC :", scores.std())
```

```python
results = []

trained_models = {}
```

```python
for name, model in models.items():

    model.fit(X_train, y_train)

    trained_models[name] = model

    pred = model.predict(X_test)

    prob = model.predict_proba(X_test)[:,1]

    results.append({

        "Model":name,

        "Accuracy":
            accuracy_score(y_test,pred),

        "Precision":
            precision_score(y_test,pred),

        "Recall":
            recall_score(y_test,pred),

        "F1":
            f1_score(y_test,pred),

        "ROC-AUC":
            roc_auc_score(y_test,prob)

    })
```

```python
results_df = pd.DataFrame(results)

results_df.sort_values(

    "ROC-AUC",

    ascending=False

)
```

```python
best_model = trained_models["XGBoost"]

best_model_name = "XGBoost"

print("Selected Model :", best_model_name)
```

```python
best_model = trained_models[best_model_name]

pred = best_model.predict(X_test)

print(

classification_report(

y_test,

pred

)

)
```

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans

from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA

from sklearn.metrics import silhouette_score
```

```python
train = pd.read_csv("train_engineered.csv")

train.head()
```

```python
X = train.drop("Bankrupt?", axis=1)

print(X.shape)
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

print("Explained Variance:")

print(pca.explained_variance_ratio_)
```

```python
inertia = []

K = range(2,11)

for k in K:

    model = KMeans(

        n_clusters=k,

        random_state=42,

        n_init=20

    )

    model.fit(X_scaled)

    inertia.append(model.inertia_)
```

```python
plt.figure(figsize=(8,5))

plt.plot(K,inertia,marker="o")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.title("Elbow Method")

plt.show()
```

```python
scores = []

for k in range(2,11):

    model = KMeans(

        n_clusters=k,

        random_state=42,

        n_init=20

    )

    labels = model.fit_predict(X_scaled)

    score = silhouette_score(

        X_scaled,

        labels

    )

    scores.append(score)

    print(

        f"K={k}:",

        round(score,4)

    )
```

```python
plt.figure(figsize=(8,5))

plt.plot(

    range(2,11),

    scores,

    marker="o"

)

plt.xlabel("Clusters")

plt.ylabel("Silhouette Score")

plt.show()
```

```python
kmeans = KMeans(

    n_clusters=4,

    random_state=42,

    n_init=20

)

clusters = kmeans.fit_predict(X_scaled)

X["Cluster"] = clusters
```

```python
plt.figure(figsize=(10,7))

plt.scatter(

    X_pca[:,0],

    X_pca[:,1],

    c=clusters,

    cmap="viridis",

    s=30

)

plt.title("Financial Risk Segmentation")

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.show()
```
